```python
scores = {1: [204,170,69],
          2: [295,194,120],
          3: [248,210,90],
          4: [246,216,106],
          5: [251,234,118],
          6: [252,253,181],
          7: [255,253,252],
          8: [255,255,233]}
```

```python
def euclidean_distance(p1, p2):
    dist = ( (p1[0]-p2[0])**2 + (p1[1]-p2[1])**2 + (p1[2]-p2[2])**2)**(1/2)
    return dist
```

```python
# Return RGB array of the darkest yellow pixel

def get_darkest_point(pic):
    min_value = [10000000000000000,[0,0,0]]
    for i in range(pic.shape[0]):
        for j in range(pic.shape[1]):
            print(i)
            dist = euclidean_distance(pic[i,j],scores[1])
            if dist < min_value[0]:
                min_value[0] = dist
                min_value[1] = pic[i,j]
            
    return min_value[1]

def get_average_rgb(pic):
    arr = np.array(pic)
    red_arr = arr[:,:,0]
    green_arr = arr[:,:,1]
    blue_arr = arr[:,:,2]
    return [np.mean(red_arr), np.mean(green_arr), np.mean(blue_arr)]
```

```python
distances = {}
# darkest_point = get_darkest_point(pic)
average_point = get_average_rgb(pic)
print(average_point)
for k,v in scores.items():
    distances[k] = euclidean_distance(average_point,v)
```
