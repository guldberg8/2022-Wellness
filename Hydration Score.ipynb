```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Defining the RGB values for each urine score

scores = {1: [204,170,69],
          2: [295,194,120],
          3: [248,210,90],
          4: [246,216,106],
          5: [251,234,118],
          6: [252,253,181],
          7: [255,253,252],
          8: [255,255,233]}
```

```python
def euclidean_distance(p1, p2):
    dist = ( (p1[0]-p2[0])**2 + (p1[1]-p2[1])**2 + (p1[2]-p2[2])**2)**(1/2)
    return dist

# Return RGB array of the average RGB of all pixels

def get_average_rgb(pic):
    arr = np.array(pic)
    red_arr = arr[:,:,0]
    green_arr = arr[:,:,1]
    blue_arr = arr[:,:,2]
    return [np.mean(red_arr), np.mean(green_arr), np.mean(blue_arr)]
```

```python
filenames = ['04-27-22.jpeg']

scores_dict = {}
for filename in filenames:
    pic = imageio.imread(filename)
    date = filename.replace('.jpeg', '')
    plt.figure(figsize = (15,15))

    plt.imshow(pic)

    print('Shape of the image : {}'.format(pic.shape))
    print('Image Hight {}'.format(pic.shape[0]))
    print('Image Width {}'.format(pic.shape[1]))
    print('Dimension of Image {}'.format(pic.ndim))
    print('Image size {}'.format(pic.size))
    print('Maximum RGB value in this image {}'.format(pic.max()))
    print('Minimum RGB value in this image {}'.format(pic.min()))

    distances = {}
    average_point = get_average_rgb(pic)
    print(f'Average RGB values for {date}: {[round(i,2) for i in average_point]}')
    for k,v in scores.items():
        distances[k] = euclidean_distance(average_point,v)
    display(distances)
    score = min(distances, key=distances.get)
    print(f'Urine Score for {date}: {score}')
    scores_dict[date] = score
    print('_____________________________________________________________________________________')
```
